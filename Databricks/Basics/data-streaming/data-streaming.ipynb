### Number(3)

Extracting a list containing all the sizes.

```
sizeList = [x.size for x in files]
```

```
print(sizeList)
```

Create a list of paths from the list you created in Point 2 called `pathList`.

```
pathList = [x.path for x in files]
```

```
print(pathList)
```

### Number(4)

using `sc.parallelize` to create an RDD from `pathList` and Naming this as `filesRDD`.

```
filesRDD = sc.parallelize(pathList)
```

### Number(5)

Printing the number of elements in this list

```
filesRDD.collect()
```

### Number (6)

Produce a `csvFilesRDD`, which only contains only those files whose names end with .csv

```
csvFilesRDD = filesRDD.map(lambda x: x.split('.'))\
                .filter(lambda y: y[-1] == 'csv')
```

```
csvFilesRDD.collect()
```

### Number(7)

Finding out how many files were removed in Point 6.

```
removedFile = filesRDD.count() - csvFilesRDD.count()
removedFile
```

### Number(8 a)

Reordering the timestamp so that we’re guaranteed the natural order is the order we want, i.e. we want to change
MM-DD-YYYY.csv to YYYYMMDD.

```
convertingRDD = csvFilesRDD.map(lambda x: (x[0]+'.'+x[1]))
```

```
convertingRDD.take(5)
```

### Number (8 b)

To stop string based sorting, the new timestamp needs to be cast to integer

```
import re
# Assumes input MM-DD-YYYY.csv
def convert_name(filepath):
    m = re.match("^.*([0-9]{2})-([0-9]{2})-([0-9]{4}).csv$", filepath)
    return (int (m.group(3) + m.group(1) + m.group(2)), filepath)
```

```
csvPairRDD = convertingRDD.map(convert_name)
```

```
csvPairRDD.take(5)
```

### Number(9)

Using `sortByKey` to find the most recent pair

```
sortedRDD = csvPairRDD.sortByKey(ascending=False)
```

```
sortedRDD.collect()
```

### Number(10)

Assigning the path of the most recent pair to the variable `chosenfile`

```
chosenfile = sortedRDD.filter(lambda p: p).top(1)
```

```
chosenfile
```

### Number(11)

Loading `chosenfile` into a Spark dataframe

```
chosenfilepath = ''
for file in chosenfile:
    chosenfilepath = file[1]
```

```
sparkDF = spark.read.csv(chosenfilepath, header='true')
```

header forms the first line of the data

```
sparkDF.printSchema()
```
