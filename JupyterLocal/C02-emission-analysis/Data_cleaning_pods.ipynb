## Replace missing values with simple random imputation regression value of each column

```python
missing_columns = ["year_1960", "year_1965", "year_1970", "year_1975", "year_1980", 
                   "year_1985", "year_1990", "year_1995", "year_2000", "year_2005", 
                   "year_2010", "year_2015", "year_2020"]
```

```python
def random_imputation(Total_Data, year):

     number_missing = Total_Data[year].isnull().sum()
     observed_values = Total_Data.loc[Total_Data[year].notnull(), year]
     Total_Data.loc[Total_Data[year].isnull(), year + '_imp'] = np.random.choice(observed_values, number_missing, replace = True)
    
     return Total_Data
```

```python
for year in missing_columns:
    Total_Data[year + '_imp'] = Total_Data[year]
    Total_Data = random_imputation(Total_Data, year)
```

```python
def outliers(Total_Data, ft):
    Q1 = Total_Data[ft].quantile(0.25)
    Q3 = Total_Data[ft].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    ls = Total_Data.index[(Total_Data[ft] < lower_bound) | (Total_Data[ft] > upper_bound)]
    
    return ls
```

```python
index_list = []
for feature in ['year_1960_imp','year_1965_imp','year_1970_imp','year_1975_imp','year_1980_imp','year_1985_imp','year_1990_imp','year_1995_imp','year_2000_imp','year_2005_imp','year_2010_imp','year_2015_imp','year_2020_imp']:
    index_list.extend(outliers(Total_Data, feature))
    
```

```python
def remove(Total_Data, ls):
    ls = sorted(set(ls))
    Total_Data = Total_Data.drop(ls)
    return Total_Data
```
